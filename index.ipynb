## Lambda Functions

```python
import pandas as pd
df = pd.read_csv('Yelp_Reviews.csv', index_col=0)
df.head(2)
```

## Simple arithmetic

Use a lambda function to create a new column called `'stars_squared'` by squaring the stars column.

```python
# Your code here
df['stars_squared'] = df['stars'].map(lambda x: x**2)
df
```

## Dates
Select the month from the date string using a lambda function.

```python
# Your code here
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].map(lambda x: x.month)#x.to_datetime.month()).head()
df.head()
```

## What is the average number of words for a yelp review?
Do this with a single line of code!

```python
# Your code here
average_words_per_review = sum(df['text'].map(lambda x: len(x.split())))/len(df['text'])
average_words_per_review
```

## Create a new column for the number of words in the review

```python
# Your code here
df['word_count'] = df['text'].map(lambda x: len(x.split()))
df[['text','word_count']].head()
```

## Rewrite the following as a lambda function

Create a new column `'Review_Length'` by applying this lambda function to the `'Review_num_words'` column. 

```python
# Rewrite the following function as a lambda function
def rewrite_as_lambda(value):
    if len(value) < 50:
        return 'Short'
    elif len(value) < 80:
        return 'Medium'
    else:
        return 'Long'
# Hint: nest your if, else conditionals


#I believe the rewrite_as_lambda function is meant to be applied to the split of the 'text' column, 
#since it deals with len(value),
#not the integer value itself.
#applying the function for checking
df['Review_length_check'] = df['text'].map(lambda x: rewrite_as_lambda(x.split()))
#df.loc[df['Review_length_check'] == 'Short',:]

#writing the lambda function
df['Review_length'] = df['word_count'].map(lambda x: 'Short' if x<50 else 'Medium' if x<80 else 'Long')
df[['Review_length_check','Review_length']]
```
